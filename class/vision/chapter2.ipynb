```python
import tensorflow as tf
import keras
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
img_rows = 28
img_cols = 28

inputshape = (img_rows, img_cols, 1)
```

```python
x_train =x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
```

```python
x_train.shape
```

```python
x_train =x_train.astype('float32') / 255.
x_test = x_test.astype('float32') /255.
```

```python
batch_size = 128
num_classes = 10
epochs = 10
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test =  keras.utils.to_categorical(y_test, num_classes)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
```

```python
model4 = Sequential()
model4.add(Conv2D(filters=10, kernel_size=5, activation="relu", input_shape=inputshape))

model4.add(MaxPooling2D(pool_size=2))
model4.add(Conv2D(filters=20, kernel_size=5, strides=1, activation="relu"))
model4.add(MaxPooling2D(pool_size=2))
model4.add(Flatten())
model4.add(Dense(4096, activation="relu"))
model4.add(Dense(10,activation="softmax"))
```

```python
model4.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model4.fit(x_train,y_train, batch_size= batch_size, epochs=epochs, verbose=1, validation_data=(x_test, y_test))
# validation_data 는 train의 일부를 train의 검증하면서 다시 학습
```

```python
score = model4.evaluate(x_test, y_test, verbose=0)
```

```python
from keras.datasets import cifar10
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
# 정규화(dataset 전처리)
X_train = X_train.astype(float) / 255.
X_test = X_test.astype(float) / 255.

# one-hot encoding


batch_size = 128
num_classes = 10
epochs = 10



y_train = keras.utils.to_categorical(y_train, num_classes)
y_test =  keras.utils.to_categorical(y_test, num_classes)
```

```python
model = Sequential() 
model.add(Conv2D(32, (3,3), input_shape=X_train.shape[1:], activation='relu', padding='same'))        
model.add(Conv2D(64, (3,3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, batch_size=64, epochs=80, validation_data=(X_test, y_test), shuffle=True)
```
