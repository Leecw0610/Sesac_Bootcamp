```python
from transformers import AutoTokenizer
```

```python
model = "matthewburke/korean_sentiment"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
raw_inputs = [
    "너희들은 안늙었구나..",
    "언론이 왜이렇게 과하게 밀어주는거지? 일본 머니가 힘이 좋기는 좋구나...",
    "기억상실증에 걸렸으면 좋겠다. 다시 보고 다시보고 다시보고 싶다."
]
```

```python
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt") #pt = pytorch
```

```python
inputs
```

```python
from transformers import AutoModel
```

```python
pre_model=AutoModel.from_pretrained(model)
```

```python
outputs=pre_model(**inputs)
```

```python
outputs.last_hidden_state.shape # input의 갯수, 문장당 token의 max-갯수, token당 출력하는 배열의 갯수
```

```python
from transformers import AutoModelForSequenceClassification
```

```python
clf = AutoModelForSequenceClassification.from_pretrained(model)
```

```python
outputs2 = clf(**inputs)
```

```python
outputs2.logits # 1,2,3문장에 대한 결괏값 (softmax 들어가기 전 값)
```

```python
import torch
```

```python
prediction = torch.nn.functional.softmax(outputs2.logits, dim=-1)
```

```python
prediction
```

```python
clf.config.id2label
```
