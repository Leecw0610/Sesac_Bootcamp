```python
import torch
from ratsnlp.nlpbook.ner import NERTrainArguments
```

```python
args = NERTrainArguments(
    pretrained_model_name = "beomi/kcbert-base",
    downstream_corpus_name = "ner",
    downstream_model_dir = "/content/drive/MyDrive/class/language/checkpoint-ner",
    batch_size=32,
    learning_rate = 5e-5,
    max_seq_length=64,
    epochs=3,
    seed=7
)
```

```python
from ratsnlp import nlpbook
```

```python
nlpbook.download_downstream_dataset(args)
```

```python
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained(
    args.pretrained_model_name,
    do_lower_case=False
)
```

```python
from ratsnlp.nlpbook.ner import NERCorpus, NERDataset
```

```python
corpus = NERCorpus(args)
```

```python
train_dataset = NERDataset(
    args = args,
    corpus = corpus,
    tokenizer = tokenizer,
    mode = 'train'
)
```

```python
train_dataset[0]
```

```python
from torch.utils.data import DataLoader, RandomSampler, SequentialSampler
```

```python
train_dataloader = DataLoader(
    train_dataset,
    batch_size=args.batch_size,
    sampler = RandomSampler(train_dataset, replacement=False),
    collate_fn = nlpbook.data_collator,
    drop_last=False,
    num_workers = args.cpu_workers
)
```

```python
val_dataset = NERDataset(
    args = args,
    corpus = corpus,
    tokenizer = tokenizer,
    mode = 'val'
)
```

```python
val_dataloader = DataLoader(
    val_dataset,
    batch_size=args.batch_size,
    sampler = SequentialSampler(val_dataset),
    collate_fn = nlpbook.data_collator,
    drop_last=False,
    num_workers = args.cpu_workers
)
```

```python
from transformers import BertConfig, BertForTokenClassification
```

```python
pretrained_model_config = BertConfig.from_pretrained(
    args.pretrained_model_name,
    num_labels = corpus.num_labels
)
```

```python
model = BertForTokenClassification.from_pretrained(
    args.pretrained_model_name,
    config=pretrained_model_config
)
```

```python
from ratsnlp.nlpbook.ner import NERTask
```

```python
task = NERTask(model, args)
```

```python
trainer = nlpbook.get_trainer(args)
```

```python
trainer.fit(task, train_dataloaders= train_dataloader, val_dataloaders=val_dataloader)
```
