```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
classifier(['I hate this so much', "It's good for health", "Is it possible?", "I agree", "only you"])
```

```python
classifier2 = pipeline("zero-shot-classification")
```

```python
classifier2("This is a python3.", candidate_labels=['snacks','books','play','computer'])
```

```python
classifier2("That's wonderful.", candidate_labels=['landscape','nature','artifact','culture','nice','cool'])
```

```python
generator = pipeline('text-generation')
```

```python
generator("Python3,", max_length=100)
```

```python
 unmasker = pipeline('fill-mask')
```

```python
ner = pipeline('ner', grouped_entities=True)
```

```python
question_answerer = pipeline("question-answering")
```

```python
question_answerer(
    context= '''In this context, a few questions are raised. The two Army barracks
that were buried in the Chorwon landslide were built just under a slope
which is prone to mudslides. Analysis of the ground is a must for
construction and military barracks are no exception.''',
    question = "Where is that Chorwon?"
)
```
