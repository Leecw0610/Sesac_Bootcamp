```python
def percent_(df_bin, df) :
    total = sum(df_bin)
    #print(total)
    percentage = df_bin/total*100
    #print(percentage)
    df['%']=percentage
    ## 데이터프레임의 항목당 백분율 계산 함수 (매개변수 (DF_백분율로 나타내려는 column, DF))
```

```python
mapping = {k:v for k, v in zip(key, value)}
mapping
```

```python
for i, k in enumerate(mapping.keys()):
    if i >= 42:
        mapping[k] = None
```
