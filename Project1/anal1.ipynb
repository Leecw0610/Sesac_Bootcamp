```python
def percent_1(col, df) :
    total = sum(df[col])
    #print(total)
    percentage = df[col]/total*100
    #print(percentage)
    df['%']=percentage
    return df
    ## 데이터프레임의 항목당 백분율 계산 함수 (매개변수 (DF_백분율로 나타내려는 column, DF))
```

```python
df_ls = []
for i in range(0, 217, 7):
    # print(i)
    df = data3_1_1.iloc[i:i+7]
    df = percent_1('데이터', df)
    df_ls.append(df)
```

```python
df_ls
```

```python
data3_2=pd.concat(df_ls)
```
