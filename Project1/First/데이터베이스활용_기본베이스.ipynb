# 데이터 베이스 활용

```python
# 해당 데이터들의 정보와 DataFrame를 불러올 수 있는 딕셔너리
# 파일 크기 큼 주의
with open('Database.p', 'rb') as file:    
    DataBase = pickle.load(file)
```

```python
# 해당 데이터들의 정보들만 빠르게 파악할 수 있는 딕셔너리
with open('Database_info.p', 'rb') as file:
    DataBase_info = pickle.load(file)
```

```python
def db_NameList():
    """데이터베이스 안에 있는 data 이름들 리스트로 반환

    Returns:
        list: data Names in DataBase
    """
    return list(DataBase_info.keys())
```

```python
def db_nameToInfo(name):
    """데이터베이스 중에 name 에 해당하는 정보들을 딕셔너리 형태로 반환

    Args:
        name (str): 데이터 이름 (한글)

    Returns:
        dict: name data's information
    """
    return DataBase_info[name]
```

```python
def db_nameToDataframe(name):
    """데이터베이스 중에 name 에 해당하는 데이터 프레임 반환

    Args:
        name (str): 데이터 이름 (한글)

    Returns:
        DataFrame: name data's DataFrame
    """
    return DataBase[name]['data'].copy()
```

```python
def db_nameToColumns(name):
    """데이터베이스 중에 name 에 해당하는 데이터 프레임의 칼럼 리스트 반환

    Args:
        name (str): 데이터 이름 (한글)

    Returns:
        list: name data's columns
    """
    return db_nameToInfo(name)['columns']
```

# 다른 유형 데이터과의 결합

```python
cob_ls = []
cross_ls = []
t = '-'
for name_x in db_NameList():
    for name_y in db_NameList():
        cob = {name_y, name_x}
        if (cob not in cob_ls) and (name_x != name_y):
            cob_ls.append(cob)
            
            x_cols = set(db_nameToColumns(name_x))
            y_cols = set(db_nameToColumns(name_y))
            
            cross = x_cols & y_cols
            if len(cross) > 2:
                print(f'[{name_x}] & [{name_y}]')
                print(f'{t*10}교집합{t*10}')
                print(cross)
                cross_ls.append(cross)
                print()
                print(f'{t*10}차집합{t*10}')
                print(f'x_cols - {x_cols - y_cols}')
                print(f'y_cols - {y_cols - x_cols}')
                print()
                print()
                print()
                print()
            
```

```python
# cross_dict 세팅
cross_dict = {}
len_cross_dict = defaultdict(set)
for cross in cross_ls:
    cross_dict[str(cross)] = {}
    cross_dict[str(cross)]['cross'] = cross
    cross_dict[str(cross)]['df'] = set()
    cross_dict[str(cross)]['len'] = len(cross)
    len_cross_dict[len(cross)].add(str(cross))
```

```python
# cross_dict 에 값 넣기
for key, value in cross_dict.items():
    cross = value['cross']
    # print(cross)
    for db_name in db_NameList():
        cols = set(db_nameToColumns(db_name))
        if len(cross - cols) == 0:
            value['df'].add(db_name)
```

```python
# 교집합이 많은 순으로
ls = list(len_cross_dict.keys())
ls.sort(reverse=True)
cob_dict = defaultdict(list)
for len_key in ls:
    print(f'{t*20}{len_key}{t*20}')
    for str_cross in len_cross_dict[len_key]:
        df_set = cross_dict[str_cross]['df']
        print(f'{str_cross}의 칼럼들을 가지고 있는 테이블')
        print(df_set)
        cob_dict[str(df_set)].append(str_cross)
        print()
        print()
        print()
```

```python
for cob, cross in cob_dict.items():
    print(cob)
    for c in cross:
        print(c)
    print()
    print()
```
