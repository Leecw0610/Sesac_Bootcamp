```python
data = np.array([0, None])
```

```python
data
```

```python
col_list = ['Glucose', 'BloodPressure' ,'SkinThickness' ,'Insulin' ,'BMI']
for i in col_list :
    for j in data[data[i]==0].index :
        np.where(data.loc[j][i]==0, None, data.loc[j][i])
```

```python
data
```

```python
data[['Glucose', 'BloodPressure' ,'SkinThickness' ,'Insulin' ,'BMI']]=data[['Glucose', 'BloodPressure' ,'SkinThickness', 'Insulin', 'BMI']].replace(0, None)
```

```python
data.isnull().sum()
```

```python
data['Pregnancies']
```

```python
data['Pregnancies2']=np.where(data['Pregnancies']==0, 0,
                     np.where(data['Pregnancies']==1, 1,
                     np.where(data['Pregnancies']==2, 2,
                     np.where(data['Pregnancies'] <5, 3,
                     np.where(data['Pregnancies'] <7, 4,
                     np.where(data['Pregnancies'] <10, 5, 6))))))      
                        
```

```python
data.isnull().sum()
```

```python
data[['Glucose', 'BloodPressure' ,'SkinThickness' ,'Insulin' ,'BMI']].fillna(data.groupby('Pregnancies2')[['Glucose', 'BloodPressure' ,'SkinThickness' ,'Insulin' ,'BMI']].transform('median'),inplace=True)
```

```python
data
```

```python
col_list=['Glucose', 'BloodPressure' ,'SkinThickness' ,'Insulin' ,'BMI']
for col in col_list :
    data[col].fillna(data.groupby('Pregnancies2')[col].transform('median'), inplace=True)
```
