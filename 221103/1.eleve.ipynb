```python
data = []
col_names=[]
for m in tqdm(range(1, 3)):
    url = 'http://apis.data.go.kr/openapi/service/ElevatorInformationService/getElevatorList'
# page1 불러오기'
    serviceKey = '3kygm0EzV%2FeG00gMCdgTQjYgzVPeoD2d32kFk9tt9zK8azyGuHSH3DDt8jWoU9vhjMqn7gbVhjnXizJ76qne8g%3D%3D'
    numOfRows = '100' 
    pageNo = str(m)
    sido = '&EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C'
    sigungu = '%EC%9D%80%ED%8F%89%EA%B5%AC'
    
    api_url = url + '?' + 'serviceKey=' + serviceKey + '&numOfRows=' \
    + numOfRows + '&pageNo=' + pageNo + '&sido=' + sido +'&sigungu=' +sigungu
    
    req = requests.get(api_url)
    soup = BeautifulSoup(req.text,'xml')
    if not col_names :
        for item in soup.items()[0]:
            col_names.append(item.name)
    for item in soup.items :
        temp = []
        for t in item :
            temp.append(t.text.strip())
        data.append(temp)
        
df = pd.DataFrame(data, columns=col_names)
```
