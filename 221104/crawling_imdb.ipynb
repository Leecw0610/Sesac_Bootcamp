```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.imdb.com/chart/boxoffice/?ref_=nv_ch_cht'
```

```python
req = requests.get(url)
html = req.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.select('#boxoffice > table > tbody > tr:nth-child(1) >td> td.titleColumn')
```

```python
title = soup.select('#boxoffice > table > tbody > tr >td >td.titleColumn >a') 
#아무것도 안나올 때 상위 구간을 지우면서 한 구간씩 찾아가기!
```

```python
data = []
for t in title :
    temp = []
    temp.append(t.text)        #1번 title 뽑아내기~
    temp.append(t['href'].replace('/title/', ''))
    data.append(temp)
```

```python
def soup_get(url) :
    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'html.parser')
    return soup
#soup을 써서 url을 따는 걸 함수로!
```
